# **Lab Task: Parking Lot Image Processing**
> This lab involves processing an image of a parking lot to detect available parking slots. The tasks include loading and displaying the image, converting it to grayscale, analyzing its dimensions, cropping a specific region, counting white spaces, and drawing bounding boxes.

# **Importing Required Libraries**

```python
# Import necessary libraries
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt
```

## **Explanation:**
> * The Pillow library is used for image processing tasks, such as opening, converting, cropping, and saving images.
> * NumPy is used for numerical computations, like analyzing pixel intensities.
> * Matplotlib is used to display the images and plots for visualization.

## **Task 1: Open and Display the Parking Lot Image**

```python
# Load the parking lot image
image_path = "parking_lot.jpg"
parking_lot_image = Image.open(image_path)

# Display the image using Matplotlib
plt.imshow(parking_lot_image)
plt.title("Original Parking Lot Image")
plt.axis("off")  # Hide axes for a cleaner display
plt.show()
```

## **Explanation:**
> * The Image.open() function loaded the parking lot image from the given path.
> * plt.imshow() displayed the image using Matplotlib. The title and axis hiding were added for a cleaner look.

This task ensured the image was loaded correctly and allowed us to visually confirm the input image before further processing.

## **Task 2: Convert to Grayscale and Save**

```python
# Convert the image to grayscale
parking_lot_gray = parking_lot_image.convert("L")

# Save the grayscale image
parking_lot_gray.save("parking_lot_gray.jpg")

# Display the saved grayscale image
plt.imshow(parking_lot_gray, cmap="gray")
plt.title("Grayscale Parking Lot Image")
plt.axis("off")
plt.show()
```

## **Explanation:**
> * The image was converted to grayscale using the convert("L") method, which removes color information, leaving only brightness levels.
> * The grayscale image was saved as parking_lot_gray.jpg for further processing.

Converting the image to grayscale simplifies processing tasks like pixel intensity analysis since we only deal with one channel instead of three (RGB).


## **Task 3: Analyze Image Dimensions**

```python

# Load the color image
color_image = Image.open('parking_lot.jpg')

# Convert the image to grayscale
gray_image = color_image.convert('L')

# Analyze dimensions for the color image
width_color, height_color = color_image.size
channels_color = len(color_image.getbands())  # Number of channels
print(f"Color Image Dimensions:\nWidth = {width_color}\nHeight = {height_color}\nChannels = {channels_color}\n -----")

# Analyze dimensions for the grayscale image
width_gray, height_gray = gray_image.size
channels_gray = len(gray_image.getbands())  # Number of channels
print(f"Grayscale Image Dimensions:\nWidth = {width_gray}\nHeight = {height_gray}\nChannels = {channels_gray}")
```

## **Explanation:**

> * The size attribute retrieved the images width and height.
> * The getbands() function determined the number of color channels (e.g., RGB has 3 channels).
> * The .size attribute retrieves the width and height for both images.
> * The .getbands() method identifies the number of channels:
>    * Color image: 3 channels (RGB).
>    * Grayscale image: 1 channel (Luminance).
> * The dimensions and channel information are printed for both images.

This task helped us understand the image's structure (size and channels), which is crucial for accurate cropping and analysis in later tasks.

## **Task 4: Crop a Parking Slot and Save**


```python
crop_coordinates = (580, 10, 880, 630)  # (left, top, right, bottom)

# Crop the image
cropped_slot = parking_lot_image.crop(crop_coordinates)

# Save the cropped section
cropped_slot.save("slot1.jpg")

# Display the saved cropped image
plt.imshow(cropped_slot)
plt.title("Cropped First Parking Slot")
plt.axis("off")
plt.show()
```

## **Explanation:**
> * Using crop(), a specific region of the image (representing a parking slot) was extracted based on the given coordinates.
> * The cropped region was saved as slot1.jpg and displayed for verification.

Cropping allows us to isolate specific areas of interest (parking slots) for focused analysis, such as detecting whether they are available

## **Task 5: Count White Spaces in Cropped Slot**


```python
# Convert the cropped image to grayscale
cropped_gray = cropped_slot.convert("L")

# Convert grayscale image to NumPy array
pixel_array = np.array(cropped_gray)

# Count white pixels (intensity = 255)
white_pixel_count = np.sum(pixel_array == 255)

# Print the count of white pixels
print(f"White Pixel Count: {white_pixel_count}")

# Display the cropped grayscale image
plt.imshow(cropped_gray, cmap="gray")
plt.title("Cropped Slot in Grayscale")
plt.axis("off")
plt.show()
```

## **Explanation:**
> * The cropped slot was converted to grayscale for intensity-based analysis.
> * Using NumPy, the pixel array was analyzed to count white pixels (intensity = 255), which might indicate empty parking spaces.


Counting white spaces helps determine parking availability. A higher count of white pixels may mean the slot is empty or less occupied.

## **Task 6: Draw Bounding Boxes Around Parking Slots**

```python
# Create a copy of the original image to draw on
highlighted_image = parking_lot_image.copy()
draw = ImageDraw.Draw(highlighted_image)

# Draw rectangles around three parking slots
draw.rectangle((100, 10, 470, 150), outline="red", width=3)
draw.rectangle((100, 190, 470, 610), outline="red", width=3)
draw.rectangle((580, 350, 740, 630), outline="red", width=3)

# Save the highlighted image
highlighted_image.save("highlighted_parking_lot.jpg")

# Display the saved highlighted image
plt.imshow(highlighted_image)
plt.title("Highlighted Parking Slots")
plt.show()
```

## **Explanation:**
> * A copy of the original image was created to draw bounding boxes around specific parking slots.
> * Red rectangles were drawn using the rectangle() method with the specified coordinates.
> * The modified image was saved as highlighted_parking_lot.jpg and displayed.


This task visually highlights the detected parking slots, providing a clear indication of the slots being analyzed.
